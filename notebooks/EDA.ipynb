```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = pd.read_csv('../input/train.csv')
```

```python
test = pd.read_csv('../input/test.csv')
```

```python
df = pd.concat([train,test],axis=0)
```

```python
df.shape
```

```python
df.drop('ID', inplace=True, axis=1)
```

```python
num = set(df.select_dtypes(include=np.number).columns)
cat = set(df.select_dtypes(exclude=np.number).columns)
```

# Univariate

```python
df_0 = df.loc[df['Loan Status'] == 0]
df_1 = df.loc[df['Loan Status'] == 1]
```

```python
for i in set(df.select_dtypes(include=np.number).columns):
    plt.plot(df_0[i], np.zeros_like(df_0[i]),'o')
    plt.plot(df_1[i], np.zeros_like(df_1[i]),'o')
    plt.xlabel(i)
    plt.show()
```
